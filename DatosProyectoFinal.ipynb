```python
NatT = NatalidadTotal.T['Total'].values
MorT = MortalidadTotal.T['Total'].values
AñosN = NatalidadTotal.columns.tolist()
df1 = pd.DataFrame({'Años': AñosN, 'Natalidad': NatT, 'Mortalidad': MorT})
df1
```

```python
# Convertir las columnas al tipo de dato numérico
df1['Años'] = pd.to_numeric(df1['Años'], errors='coerce')
df1['Natalidad'] = pd.to_numeric(df1['Natalidad'], errors='coerce')
df1['Mortalidad'] = pd.to_numeric(df1['Mortalidad'], errors='coerce')

# Filtrar las filas que contienen valores no numéricos (NaN)
df1 = df1.dropna()
```

```python
# Definir las variables independientes (X) y dependientes (y)
X1 = df1[['Natalidad', 'Mortalidad']]
y1 = df1[['Años']]

# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=0)

# Crear el modelo de regresión lineal
model1 = LinearRegression()

# Ajustar el modelo a los datos de entrenamiento
model1.fit(X_train, y_train)

# Predecir los valores para el conjunto de prueba
y_pred = model1.predict(X_test)

# Obtener los coeficientes de la regresión
coef = model1.coef_
intercept = model1.intercept_

# Imprimir los resultados
print('Coeficientes:', coef)
print('Intercepto:', intercept)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

N = X1['Natalidad']
M = X1['Mortalidad']
target = y1

def plano(x1, x2):
    m1 = model1.coef_[0][0]
    m2 = model1.coef_[0][1]
    b = model1.intercept_[0]
    return m1*x1 + m2*x2 + b

X1 = np.linspace(np.min(N), np.max(N), 20)
X2 = np.linspace(np.min(M), np.max(M), 20)

x1grid, x2grid = np.meshgrid(X1, X2)
Z = plano(x1grid, x2grid)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X1, X2, Z, linewidth=1, color='lime', alpha=0.3)
ax.scatter(N, M, target, marker='o', s=30, color = 'b')
ax.scatter(X_test['Natalidad'], X_test['Mortalidad'], y_pred, '^', color='orange')
ax.set_xlabel('Natalidad')
ax.set_ylabel('Mortalidad')
ax.set_zlabel('Años')
#ax.view_init(0, 45)
plt.show()
```
