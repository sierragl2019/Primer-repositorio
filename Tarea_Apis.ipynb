```python
import requests
import pandas as pd
import numpy as np
```

```python
palabra = 'astro'
 
endpoint = 'https://es.wikipedia.org/w/rest.php/v1/search/title'

params = {
            'q' : palabra,
            'limit': 20
        }

request5_wiki = requests.get(endpoint, params=params)

request5_wiki.status_code
```

```python
datos = pd.json_normalize(request5_wiki.json()['pages'])
```

```python
Resumenes=list()
titulo=list()
pagina=list()
for i in range(len(datos)):
  tit=datos.iloc[i]['title']
  pag=datos.iloc[i]['id']

  titulo.append(tit)
  pagina.append(pag)

  endpoint = 'https://es.wikipedia.org/w/api.php'

  params = {
            'action' : 'query',
            'format' : 'json',
            'titles' : tit, 
            'prop' : 'extracts',
            'exintro': True,
            'explaintext': True
          }

  request3_wiki = requests.get(endpoint, params=params)
  request3_wiki.status_code
  res = pd.json_normalize(request3_wiki.json()['query'])
  Res= res[f'pages.{pag}.extract'][0]
  Resumenes.append(Res)
```

```python
contiene=list()
for i in range(len(Resumenes)):
  if 'ciencia' in Resumenes[i] or 'física' in Resumenes[i]:
        contiene.append(1)
  else:
        contiene.append(0)
print(contiene)
```

```python
DR = { 'page_id' : pagina,'titulo' : titulo, 'resumen' : Resumenes, 'contiene_palabra' : contiene}
df = pd.DataFrame(DR)
df
```
